```python
import pandas as pd


df = pd.read_csv('data/cause_of_death_indonesia_clean.csv')

df
```

```python
df = df[(df['Type'] == 'Bencana Non Alam dan Penyakit') & (~df['Cause'].str.contains('kecelakaan', case=False, regex=False)) & (~df['Cause'].str.contains('kebakaran', case=False, regex=False)) & (~df['Cause'].str.contains('teknologi', case=False, regex=False))].reset_index()

df
```

```python
df = df[df['Total Deaths'] != 0]

df
```

```python
df.loc[df['Cause'].str.startswith('KLB '), 'Cause'] = df.loc[df['Cause'].str.startswith('KLB'), 'Cause'].str.replace('KLB ', '')
```

```python
antraks = df[df['Cause'].str.contains('Antraks', case=False)]

new_antraks = antraks.groupby('Year')['Total Deaths'].sum().reset_index()

for index, row in new_antraks.iterrows():
   year = row['Year']
   total_deaths = row['Total Deaths']
   df.loc[(df['Cause'].str.contains('Antraks', case=False)) & (df['Year'] == year), 'Total Deaths'] = total_deaths

df.loc[df['Cause'].str.contains('Antraks', case=False), 'Cause'] = 'Antraks'

```

```python
infectious_disease = ['AIDS', 'Antraks', 'Avian Influenza (Flu Burung)', 'Batuk Rejan', 'Campak', 'Cikungunya', 'COVID-19', 'Demam Tifoid dan Paratifoid', 
                     'Difteri', 'Gastroenteritis', 'Hepatitis A', 'Hepatitis B', 'Malaria', 'Leptospirosis', 'Rabies', 'Rubella', 'Tuberkulosis', 'Tetanus', 
                     'Demam Berdarah', 'Disentri', 'Frambusia', 'Malaria', 'Meningitis', 'Parotitis', 'Pertusis', 'Polio', 'Thypoid', 'Tifus Perut', 'Kolera', 'Leptospirosis',
                     'Penyakit Infeksi & Parasit Tertentu', 'Penyakit infeksi dan parasit kongenital', 'Penyakit Sistem Cerna', 'Penyakti Sistem Kemih Kelamin', 'Penyakit Sistem Napas',
                     'Rabies', 'Rubella', 'Tuberkulosis', 'Wabah Penyakit (Epidemi)'
                     ]

def kategori_penyakit(penyakit):
   if penyakit in infectious_disease:
      return 'Penyakit Menular'
   else:
      return 'Penyakit Lainnya'
   


df_pivot['Category'] = [kategori_penyakit(p) for p in df_pivot['Cause']]

df_pivot
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Group dan hitung total kematian berdasarkan kategori
per_category = df_pivot.groupby(['Category'])['Total Deaths'].sum()

# Plot donut chart
plt.figure(figsize=(8, 6))
sns.set_palette('pastel')
plt.pie(per_category, labels=per_category.index, autopct='%1.0f%%', startangle=90)

# Buat lingkaran di tengah untuk membuat donut chart
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Atur aspek ratio menjadi equal untuk menjaga lingkaran menjadi bentuk donut
plt.axis('equal')
plt.title('Total Kematian Berdasarkan Kategori', fontsize=15)
plt.show()
```

```python
df_pivot_infectious_disease = df_pivot[df_pivot['Category'] == 'Penyakit Menular']

df_pivot_infectious_disease
```

```python
direct_contact = ['AIDS', 'Hepatitis A', 'Hepatitis B', 'Disentri', 'Frambusia', 'Meningitis', 'Pertusis', 'Penyakit infeksi dan parasit kongenital', 'Penyakit Sistem Kemih Kelamin', 'Tetanus']
food = ['Demam Tifoid dan Paratifoid', 'Gastroenteritis', 'Leptospirosis', 'Thypoid', 'Tifus Perut', 'Kolera', 'Penyakit Sistem Cerna','Polio']
droplet = [ 'Antraks', 'Avian Influenza (Flu Burung)', 'Batuk Rejan', 'Campak', 'COVID-19', 'Difteri', 'Influenza', 'Parotitis', 'Penyakit Infeksi & Parasit Tertentu', 'Penyakit Sistem Napas', 'Rubella', 'Tuberkulosis']
vektor = ['Cikungunya', 'Demam Berdarah', 'Malaria', 'Rabies']

def penyebaran_penyakit(penyakit):
   if penyakit in direct_contact:
      return 'Kontak Langsung'
   elif penyakit in food:
      return 'Makanan'
   elif penyakit in droplet:
      return 'Droplet'
   elif penyakit in vektor:
      return 'Vektor'
   


df_pivot_infectious_disease['Jenis Penyebaran'] = [penyebaran_penyakit(p) for p in df_pivot_infectious_disease['Cause']]

df_pivot_infectious_disease
```

```python
per_penyebaran = df_pivot_infectious_disease.groupby(['Jenis Penyebaran'])['Total Deaths'].sum()

# Plot pie chart
plt.figure(figsize=(8, 6))
sns.set_palette('pastel')
plt.pie(per_penyebaran, labels=per_penyebaran.index, autopct='%1.2f%%', startangle=90)

# Atur aspek ratio menjadi equal untuk menjaga pie chart menjadi lingkaran sempurna
plt.axis('equal')

# Tambahkan judul
plt.title('Total Kematian Penyakit Menular Berdasarkan Jenis Penyebaran', fontsize=15)

# Tampilkan plot
plt.show()
```
